# Generate Test Data

```python
from __future__ import division
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import lognormal_model as ln
from helpers import *
%matplotlib inline

np.random.seed(11739)
```

## Define model parameters

```python
U = 100  # units
T = 10000  # time points/frames
Kdata = 3  # number of latent states
dt = 1/30  # seconds per frame
```

## Generate Markov chains

```python
# make transition matrix
A = np.array([[0.98, 0.05], [0.02, 0.95]])
```

```python
chain = np.empty((T, Kdata))

chain[0, :] = 0  # start all categories "off"
for t in xrange(1, T):
    for k in xrange(Kdata):
        chain[t, k] = stats.bernoulli.rvs(A[1, chain[t - 1, k]])
```

```python
matshow(chain.T);
```

## Baseline distributions

Scipy parameterizes the log-normal as
$$
\text{log-}\mathcal{N}(x) = \frac{1}{x\sqrt{2\pi \sigma^2}} \exp\left({-\frac{(\log (x/x_0))^2}{2\sigma^2}}\right)
$$ 
from which we can see that $\log x$ is normally distributed with mean $\mu = \log x_0$ and variance $\sigma^2$.

In scipy, $x_0$ is the scale parameter and $\sigma$ is the shape.

```python
bl_scale = 10
bl_mean = np.log(bl_scale)
bl_sig = 0.5
# baselines should follow a different distribution
bl = stats.norm.rvs(scale=bl_sig, loc=bl_mean, size=U) 

xx = np.linspace(0, 50, 1000)
plt.plot(xx, stats.lognorm.pdf(xx, scale=bl_scale, s=bl_sig))
plt.title('Baseline firing distribution');
```

## Firing rate effect distributions

```python
fr_sig = 0.5
fr_scale = 0.5
fr_mean = np.log(fr_scale)
lam = stats.norm.rvs(loc=fr_mean, scale=fr_sig, size=(Kdata, U))

xx = np.linspace(0, 5, 1000)
plt.plot(xx, stats.lognorm.pdf(xx, scale=fr_scale, s=fr_sig))
plt.title('Category effect firing distribution');
```

## Overdispersion

```python
if overdisp:
    xx = np.linspace(0, 2, 300)
    od_scale = 1
    od_mean = np.log(od_scale)
    od_sig = 0.10
    plt.plot(xx, stats.lognorm.pdf(xx, s=od_sig, scale=od_scale))
    plt.title('Overdispersion Effect');
```

## External regressors

```python
# just as above, make a set of Markov-esque regressors

# number of regressors
R = 3  

# "transition matrix" -- want regressors to have some temporal coherence
AX = np.array([[0.98, 0.05], [0.02, 0.95]])
```

```python
Xchain = np.empty((T, R))

Xchain[0, :] = 0  # start second category off
for t in xrange(1, T):
    for k in xrange(0, R):
        Xchain[t, k] = stats.bernoulli.rvs(AX[1, Xchain[t - 1, k]])
```

```python
matshow(Xchain.T)
```

```python
Xf = pd.DataFrame(Xchain, columns=map(lambda x: 'X' + str(x), xrange(R)))
Xf.index.name = 'frame'
Xf = Xf.reset_index()
```

```python
Xf.head()
```

```python
Xfr_scale = 1
Xfr_mean = np.log(Xfr_scale)
Xfr_sig = 0.25
Xlam = stats.norm.rvs(loc=Xfr_mean, scale=Xfr_sig, size=(R, U))

xx = np.linspace(0, 5, 1000)
plt.plot(xx, stats.lognorm.pdf(xx, s=Xfr_sig, scale=Xfr_scale))
plt.title('Regressor effect firing distribution');
```

## Calculate firing rates

Let $\lambda_0$ be the baseline effect, $b_k$ the firing rate effect for latent category $k$ with binary indicator $z_{tk}$ at time $t$, and $\beta_r$ the firing rate effect for external regressor $r$ with value $x_{tr}$ at time $t$. Finally, let $\epsilon_t$ be the overdispersion random effect. Then
$$
\eta_{tu} = \log f_{tu} = \lambda_{0u} + \sum_k z_{tk} b_{ku} + \sum_r x_{tr} \beta_{ru} + \epsilon_t
$$

```python
# add overdispersion
if overdisp:
    eps = stats.norm.rvs(loc=od_mean, scale=od_sig, size=fr.shape)
else:
    eps = 0
    
# calculate firing rate
fr = np.exp(bl + chain.dot(lam) + Xchain.dot(Xlam) + eps) * dt
```

```python
# get a crude sense of population firing
plt.hist(fr.ravel() / dt, bins=1000);
plt.xlim(0, 50);
```

# Set up Priors

## Priors on baseline:

We will use a simple model for baseline priors
$$
\lambda_{0u} \sim \mathcal{N}(m_0, v_0^2)
$$
where the actual baseline firing rate is $f = e^{\lambda_0}$.

```python
bl_prior_mean = np.log(15) + np.log(dt)
bl_prior_std = 0.7
bl_prior_prec = 1. / bl_prior_std ** 2

baseline_dict = ({'prior_mean': bl_prior_mean, 'prior_prec': bl_prior_prec})

xx = np.linspace(0, 50, 1000)
samples = np.exp(stats.norm.rvs(loc=bl_prior_mean, scale=bl_prior_std, size=1e5))
sns.kdeplot(samples / dt, clip=(0, 150))
plt.title('baseline prior distribution')
plt.xlim(0, 100);
plt.xlabel('spikes per second');
```

## Priors on latent firing rate effects:

Here, we adopt a hierarchical prior
$$
b_{ku} \sim \mathcal{N}(0, v^2_k) \\
\tau_k = \frac{1}{v^2_k} \sim \mathrm{Ga}(s, r)
$$

```python
# set number of categories to fit
K = 5
```

```python
fr_prior_prec_shape = 1
fr_prior_prec_rate = 1e-3

fr_latent_dict = ({'prior_prec_shape': fr_prior_prec_shape, 
                   'prior_prec_rate': fr_prior_prec_rate})

# make some plots
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
xx = np.linspace(0, 2, 500)
plt.plot(xx, stats.invgamma.pdf(xx, a=fr_prior_prec_shape, scale=fr_prior_prec_rate))
plt.title('latent fr effect variance prior');

plt.subplot(1, 2, 2)
var_samples = stats.invgamma.rvs(a=fr_prior_prec_shape, scale=fr_prior_prec_rate, size=1e5)
samples = stats.norm.rvs(scale=np.sqrt(var_samples), size=var_samples.shape)
sns.kdeplot(samples, clip=(-1, 1))
plt.title('latent fr effect prior');
```

## Priors on $A$, $\pi$

```python
###### A ###############
A_off = 10.
A_on = 1.
Avec = np.r_[A_off, A_on].reshape(2, 1, 1)
A_prior = np.tile(Avec, (1, 2, K))

###### pi ###############
pi_off = 15.
pi_on = 1.
pi_prior = np.tile(np.r_[pi_off, pi_on].reshape(2, 1), (1, K))

latent_dict = {'A_prior': A_prior, 'pi_prior': pi_prior}
```

```python
xx = np.linspace(0, 1, 100)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.plot(xx, stats.beta.pdf(xx, A_prior[1, 0, 0], A_prior[0, 0, 0]))
plt.title(r'Transition $0 \rightarrow 1$');

plt.subplot(1, 3, 2)
plt.plot(xx, stats.beta.pdf(xx, A_prior[1, 1, 0], A_prior[0, 1, 0]))
plt.title(r'Transition $1 \rightarrow 1$');

plt.subplot(1, 3, 3)
plt.plot(xx, stats.beta.pdf(xx, pi_prior[1, 0], pi_prior[0, 0]))
plt.title('Initial state value');
```

## Priors on $\theta$ (overdispersion)

```python
od_prior_mean = 0
od_prior_std = 0.3
od_prior_prec = 1. / od_prior_std ** 2
od_dict = {'prior_mean': od_prior_mean * np.ones((M,)), 'prior_prec': od_prior_prec * np.ones((M,))}

samples = stats.norm.rvs(loc=od_prior_mean, scale=od_prior_std, size=1e5)
sns.kdeplot(np.exp(samples), clip=(0, 5))
plt.xlim(0, 4)
plt.title('Overdispersion effect prior');
```

## Priors on $\upsilon$ (regression coefficients)

```python
ups_shape = 11.
ups_rate = 10.
reg_shape = ups_shape * np.ones((U, R))  # shape
reg_rate = ups_rate * np.ones((U, R))  # rate

reg_dict = {'prior_shape': reg_shape, 'prior_rate': reg_rate}
```

```python
xx = np.linspace(0, 2, 100)
plt.plot(xx, stats.gamma.pdf(xx, a=ups_shape, scale=1./ups_rate))
plt.title('Regression effect prior');
```

# Initial guesses for posterior parameters

```python
# baseline posteriors
bld = ({'post_shape_shape': bl_shape_shape, 'post_shape_rate': bl_shape_rate, 
        'post_mean_shape': bl_mean_shape, 'post_mean_rate': bl_mean_rate,
        'post_child_shape': np.ones((U,)), 'post_child_rate': np.ones((U,))})
baseline_dict.update(bld)
```

```python
# firing rate effects for latents
frd = ({'post_shape_shape': fr_shape_shape, 'post_shape_rate': fr_shape_rate, 
        'post_mean_shape': fr_mean_shape, 'post_mean_rate': fr_mean_rate,
        'post_child_shape': np.ones((U, K)), 'post_child_rate': np.ones((U, K))})
fr_latent_dict.update(frd)
```

```python
# latent states

# E[z]
# initialize pretty much at random (10% 1's)
rand_frac = 0.1
xi_mat = (rand_frac >= np.random.rand(T, K))
xi_mat = xi_mat.astype('float')
z_prior = np.dstack([1 - xi_mat, xi_mat]).transpose((2, 0, 1))

# E[zz]
Xi_mat = rand_frac >= np.random.rand(2, 2, T - 1, K)
Xi_mat = Xi_mat.astype('float')

ld = ({'A_post': A_prior, 'pi_post': pi_prior, 'z_init': z_prior,
       'zz_init': Xi_mat, 'logZ_init': np.zeros((K,))})
latent_dict.update(ld)
```

```python
# firing rate regressors
# since we know exact update for a_mat, use that
nn = df['count']
uu = df['unit']
NX = nn[:, np.newaxis] * df.iloc[:, -R:]
a_mat = NX.groupby(uu).sum().values
b_mat = a_mat.copy()

reg_dict.update({'post_shape': a_mat, 'post_rate': b_mat})
```

```python
# overdispersion
od_dict.update({'post_shape': np.ones((M,)), 'post_rate': np.ones((M,))})
```

# Do inference with random restarts:

```python
numstarts = 1
```

```python
fitobjs = []
Lvals = []
for idx in xrange(numstarts):
    gpm = gp.GammaModel(df, K)
    gpm.initialize_baseline(**jitter_inits(baseline_dict, 0.25))
    gpm.initialize_fr_latents(**jitter_inits(fr_latent_dict, 0.25))
    gpm.initialize_latents(**jitter_inits(latent_dict, 0.05))
    gpm.initialize_fr_regressors(**jitter_inits(reg_dict, 0.05))
    if overdisp:
        gpm.initialize_overdispersion(**jitter_inits(od_dict, 0.25))
    gpm.finalize()
    
    print "Start {} -----------------------".format(idx)
    %time gpm.do_inference(tol=1e-4, verbosity=2)
    print "Final L = {}".format(gpm.L())
    Lvals.append(gpm.L())
    fitobjs.append(gpm)
```

```python
# pick out best fit
bestind = np.argmax(Lvals)
gpm = fitobjs[bestind]
del fitobjs  # to save memory
```

## Calculate overlap between actual and inferred latents

Define the overlap as the maximum of the match or mismatch percentage. That is, if $x$ and $y$ are binary vectors, 
$$
\mathrm{overlap} = \max \{ \mathbb{E}[xy + (1-x)(1-y)], \mathbb{E}[(1-x)y + x(1-y)]\}
$$

```python
matshow(overlap, vmin=0, vmax=1)
plt.colorbar();
plt.xlabel('Inferred categories');
plt.ylabel('Actual categories');
```
