# Model 1 Scratch

Variational inference for category regressors generated by Markov chains 

```python
from __future__ import division
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

## Let's make some fake data:

```python
U = 120  # units
T = 100000  # time points/frames
K = 5  # categories
dt = 1 / 30  # frames per second
```

```python
# 0 -> 1 transition should be small
v0_a = 1
v0_b = 100
xx = np.linspace(0, 1, 1000)
plt.plot(xx, stats.beta.pdf(xx, v0_a, v0_b));

# 1 -> 1 transition should be large
v1_a = 100
v1_b = 1
xx = np.linspace(0, 1, 1000)
plt.plot(xx, stats.beta.pdf(xx, v1_a, v1_b));
```

```python
# make matrix of transition probabilities
# each row is a category, each column the transition
# probability from that state to state 1
# (this is easy to do without matrices since we have
# only two states)
v0 = stats.beta.rvs(v0_a, v0_b, size=K)
v1 = stats.beta.rvs(v1_a, v1_b, size=K)
v = np.vstack([v0, v1]).T 
```

```python
# pick initial states for each category
z0_probs = stats.beta.rvs(50, 50, size=K)
z0 = stats.bernoulli.rvs(z0_probs)
```

```python
# make the chain by evolving each category through time
chain = np.empty((K, T), dtype='int')

# initialize
chain[:, 0] = z0

for t in xrange(1, T):
    p_trans_to_1 = v[range(K), chain[:, t - 1]]
    chain[:, t] = stats.bernoulli.rvs(p_trans_to_1)
    
```

```python
# plot the categories through time
Trng = (0, 500)
plt.matshow(chain[:, slic, aspect='auto', cmap=cm.gray);
```

```python
# set up firing rates for each category, unit
lam = stats.gamma.rvs(a=10 / 3, scale=3, size=(K, U))
```

```python
# calculate rate for each time
fr = chain.T.dot(lam) * dt
fr = fr + 1e-5  # in case we get exactly 0

# draw from Poisson
N = stats.poisson.rvs(fr)
```

## Priors

### $\lambda$

```python
amat = np.ones((K, U))
bmat = 0.25 * np.ones((K, U))
```

### $z_0$

```python
zeta = 0.5 * np.ones((K,))
```

### $\nu$

Here, we choose weakly informative priors based on the assumption that $\nu_{0k} = \nu_{0 \rightarrow 1}$ is small and $\nu_{1k} = \nu_{1 \rightarrow 1}$ is large.

```python
alpha_mat = np.r_[np.ones((1, K)), 10 * np.ones((1, K))]
beta_mat = np.r_[10 * np.ones((1, K)), np.ones((1, K))]
```

```python
xx = np.linspace(0, 1, 1000)
plt.plot(xx, stats.beta.pdf(xx, 1, 10));
plt.plot(xx, stats.beta.pdf(xx, 10, 1));
```

## Initial values

### $q(\lambda)$

```python
cmat = np.ones((K, U))
dmat = 0.25 * np.ones((K, U))
```

### $q(\nu)$

```python
gamma_mat = np.r_[np.ones((1, K)), 10 * np.ones((1, K))]
delta_mat = np.r_[10 * np.ones((1, K)), np.ones((1, K))]
```

### $q(z)$

```python
xi_mat = stats.beta.rvs(50, 50, size=(T, K))
```

### $q(n)$

```python
pi_mat = np.random.rand(K, T, U)
pi_mat = pi_mat / np.sum(pi_mat, 0)
```

## Inference

```python
from scipy.special import logit, digamma
logistic = stats.logistic.cdf
```

```python
# tuple of initial values
pars0 = (cmat, dmat, gamma_mat, delta_mat, xi_mat, pi_mat)
```

```python
def update(pars):
    cmat, dmat, gamma_mat, delta_mat, xi_mat, pi_mat = pars
    cmat = amat + np.sum(N * pi_mat, 1)
    dmat = bmat + np.expand_dims(np.sum(xi_mat, axis=0), 1)
    eps = 1e-5
    log_pi = ( ((1 - xi_mat.T) * np.log(eps))[:, :, np.newaxis] + 
              xi_mat.T[:, :, np.newaxis] * (digamma(cmat) - np.log(dmat))[:, np.newaxis, :] )
    pi_mat = np.exp(log_pi)
    pi_mat = pi_mat / np.sum(pi_mat, 0)
    
    # prefactor: probability of n = 0
    log_pf = np.sum(N * np.log(1 - pi_mat), 2).T

    # create versions of xi one timestep ahead and one behind
    xi_prev = np.roll(xi_mat, 1)
    xi_next = np.roll(xi_mat, -1)

    # set boundary conditions to zeta
    xi_prev[0, :] = zeta
    xi_next[-1, :] = zeta

    # update rule
    lam_terms = -np.sum(digamma(cmat) - np.log(dmat), 1)[np.newaxis, :]
    nu_terms = (xi_prev + xi_next) * (digamma(gamma_mat[1, :]) - 
                                      digamma(delta_mat[1, :]) -
                                      digamma(gamma_mat[0, :]) +
                                      digamma(delta_mat[0, :]))
    cons_terms = ( digamma(gamma_mat[0, :]) - 
                  2 * digamma(delta_mat[0, :]) + 
                  digamma(delta_mat[1, :]) )[np.newaxis, :]
    logit_xi = np.exp(log_pf) * (lam_terms + nu_terms + cons_terms)
    xi_mat = logistic(logit_xi)
    
    # nu updates
    gamma_mat = np.sum(xi_mat * np.array([1 - xi_prev, xi_prev]), 1) + alpha_mat
    delta_mat = np.sum((1 - xi_mat) * np.array([1 - xi_prev, xi_prev]), 1) + beta_mat
    
    return (cmat, dmat, gamma_mat, delta_mat, xi_mat, pi_mat)
```

```python
pars = update(pars0)
```

```python
for ind in xrange(10):
    pars = update(pars)
```
